```python
score = int(input("Enter your grade: "))
if score >= 90:
    print("Excellent!")
elif score >= 80:
    print("Good job!")    
elif score >= 70:
    print("You can do better")    
elif score >= 60:
    print("You need to work harder") 
else:
    print("You need to improve your ferformance.")

print("Thank you for using the grade evaluator")
           
```

```python
score = int(input("Enter your score: "))
oneDigit = score // 10
match oneDigit:
    case 10 | 9 :
        print("Excellent!")
    case 8:
        print("Good job!")     
    case 7:
        print("You can do better")     
    case 6:
        print("You need to work harder")     
    case _:
        print("You need to improve your preformance. ")

print("Thank you for using the grade evaluator .")             

```

```python
# 합계를 저장할 변수
hap = 0

# 1부터 100까지 반복
for i in range(1, 101):

    # 현재 숫자를 합계에 더한다.
    hap = hap + i

# 반복이 모두 끝난 후 출력
print("1부터 100까지의 합 =", hap)
```

```python
# 홀수의 합을 저장할 변수
hap = 0

# 1부터 100까지 반복
for i in range(1, 101):

    # i를 2로 나누었을 때 나머지가 1이면 홀수이다.
    if i % 2 == 1:

        # 홀수만 합계에 더한다.
        hap = hap + i

# 결과 출력
print("1부터 100까지 홀수의 합 =", hap)
```

```python
# 6의 배수의 합을 저장할 변수
hap = 0

# 1부터 100까지 반복
for i in range(1, 101):

    # i를 6으로 나누었을 때 나머지가 0이면
    # 6의 배수이다.
    if i % 6 == 0:

        # 6의 배수만 더한다.
        hap = hap + i

# 결과 출력
print("1부터 100까지 6의 배수의 합 =", hap)
```

```python
number = int(input("Enter a number: "))

# 1. 위쪽 삼각형 그리기
for i in range(1, number + 1):
    for j in range(number - i):
        print(" ", end="")
    for j in range(2 * i - 1):
        print("*", end="")
    print()

# 2. 아래쪽 뒤집힌 삼각형 그리기
for i in range(number - 1, 0, -1):
    for j in range(number - i):
        print(" ", end="")
    for j in range(2 * i - 1):
        print("*", end="")
    print()
```
